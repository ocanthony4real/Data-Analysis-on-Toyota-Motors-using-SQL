```python
import pandas as pd
import yfinance as yf
from sqlalchemy import create_engine
```

```python
TM = yf.Ticker("TM")
```

```python
cashflow = TM.cashflow
cashflow = cashflow.transpose()
cashflow.columns = cashflow.columns.str.replace(' ', '_')
cashflow.index.names = ['Date']
cashflow.head(3)
```

```python
engine = create_engine("mysql+pymysql://root:12345@localhost/toyotamotors")
cashflow.to_sql("cashflow", con = engine)
```

```python
balance_sheet = TM.balance_sheet
balance_sheet.head(3)
```

```python
balance_sheet = balance_sheet.transpose()
balance_sheet.columns = balance_sheet.columns.str.replace(" ", "_")
balance_sheet.index.names = ['Date']
balance_sheet.columns.names = ['']
balance_sheet.head(3)
```

```python
engine = create_engine("mysql+pymysql://root:12345@localhost/toyotamotors")
balance_sheet.to_sql("balance_sheet", con = engine)
```
